```python
import pandas as pd
import numpy as np
from sklearn.cluster import MiniBatchKMeans
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('../../data/clean/final.csv')

df['label'].value_counts()
```

```python
tfidf = TfidfVectorizer(ngram_range=(1,1))
km = MiniBatchKMeans(n_clusters=12, n_init=2)
pos_df = df[df['label'] == 1]

feat = tfidf.fit_transform(pos_df['comment'])
labels = km.fit_predict(feat, pos_df['label'])

labels.shape
```

```python
pos_df['cluster'] = labels
pos_df
```

```python
pos_df.groupby('cluster').apply(lambda x: x.sample(300, weights=x['distance_to_centroid'])).reset_index(drop=True).loc[:, ~df.columns.str.contains('^Unnamed')].to_csv('../../data/clean/clustered.csv', index=False, encoding='utf-8')
```

```python
dis = km.transform(feat)
pos_df['distance_to_centroid'] = dis.min(axis=-1)
pos_df
```

```python
dis = km.transform(feat)

idxs = []

top_k = 300

for i in range(len(km.cluster_centers_)):
    d_to_i = dis[:, i]
    ind = np.argsort(d_to_i)[::-1][:top_k]
    idxs.extend(ind)

pd.concat([df.iloc[idxs], df[df['label'] == 0]]).loc[:, ~df.columns.str.contains('^Unnamed')].to_csv('../../data/clean/clustered.csv', index=False, encoding='utf-8')
```
